```python
# STEP 1: Import Libraries and Load Data 
# Code adapted from Iqbal, M. (2026) 'Class_Demo_DT.ipynb', 
# Machine Learning module, CCT College Dublin.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the abalone dataset
data = pd.read_csv('abalone.csv')

# Displaying the first 10 rows
print( "First 5 rows of the dataset")
print(data.head(10))
```

```python
# Checking the shape
print(" Dataset Shape ")
print(f"Rows: {data.shape[0]}, Columns: {data.shape[1]}")
```

```python
#  STEP 2: Check for Missing Values and Duplicates 
# Code adapted from Iqbal, M. (2026) 'Class_Demo_DT.ipynb'

print("Missing values:")
print(data.isnull().sum())
```

```python
duplicates = data.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
```

```python
# STEP 3: Remove null values and duplicates
```

```python
#I'll drop the null values and check
data = data.dropna()
print(data.isnull().sum())
```

```python
data = data.drop_duplicates()
print(data.shape)
```
